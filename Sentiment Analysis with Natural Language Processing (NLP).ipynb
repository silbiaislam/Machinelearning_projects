####  Analyzing sentiment in text data, e.g., classifying movie reviews as positive or negative.

This code performs sentiment analysis on a list of sample sentences using the Natural Language Toolkit (NLTK) library's VADER sentiment analysis tool. It computes sentiment scores (positive, negative, neutral, and compound) for each sentence, and based on the compound score, it classifies the sentences as positive, negative, or neutral. 

The sentiment analysis results are displayed for each sentence. Additionally, a bar chart is generated to visually represent the compound sentiment scores, providing an overview of the sentiment of each sentence. 

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
```

```python
# nltk.download('vader_lexicon')
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
# List of sample sentences for sentiment analysis
sentences = [
    "I love this product. It's amazing!",
    "The weather is gloomy and depressing.",
    "This movie is so boring and uninteresting.",
    "The concert was okay, but the music was too loud.",
    "I feel indifferent about this book.",
]
```

```python
# Initialize lists to store sentiment scores
compound_scores = []
positive_scores = []
negative_scores = []
neutral_scores = []
```

```python
# Analyze sentiment for each sentence
for sentence in sentences:
    sentiment = analyzer.polarity_scores(sentence)
    compound_scores.append(sentiment['compound'])
    positive_scores.append(sentiment['pos'])
    negative_scores.append(sentiment['neg'])
    neutral_scores.append(sentiment['neu'])
    print(f"Sentence: '{sentence}'")
    print("Sentiment Scores:", sentiment)
    print()
```

```python
# Create a bar chart to visualize compound scores
plt.figure(figsize=(10, 6))
plt.bar(range(len(sentences)), compound_scores, tick_label=sentences, color='c')
plt.xlabel("Sentences")
plt.ylabel("Compound Sentiment Score")
plt.title("Sentiment Analysis of Sentences")
plt.xticks(rotation=45, ha="right")

plt.show()
```
